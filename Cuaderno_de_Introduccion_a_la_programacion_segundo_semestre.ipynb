## Promedio de 4 numeros

### Diagrama

Promedio.png

### Pseudocodigo



```
Algoritmo PromediodeNumerosEnteros
	Leer N1
	Leer N2
	Leer N3
	Leer N4
	Result = (N1+N2+N3+N4)/4
	Escribir Result
```

### Codigo

## Edad Para Votar

###  Diagrama

Votar.png

### Pseudocodigo



```
Algoritmo Votar
	Leer Edad
	Si Edad > 18 Entonces
		Escribir "Si puede votar"
	SiNo
		Escribir "No puede votar"
	FinSi
FinAlgoritmo
```

## Descuento de un traje

## Diagrama

Descuento.png

### Pseudocodigo



```
Algoritmo Descuento
	Leer Traje
	Si Traje > 2500 Entonces
		Escribir "Se le aplica un 15% de descuento"
	SiNo
		Escribir "Se le aplica un 8% de descuento"
```

### Codigo

````python
def Descuento():
    Traje = int(input())

    if Traje >= 2500:
        print ("se le aplica un 15% de descuento")

    else:
        print ("Se le aplica un 8% de descuento")

Descuento()
````
